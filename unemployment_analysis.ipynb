## Data Cleaning and Preparation

Before plotting anything, we clean the data step by step. Each small cell below shows one simple operation so the notebook is easier to follow for beginners.


```python
# Step 1: standardize the column names.
df.columns = df.columns.str.strip() 
# strip is used to removes spaces from the beginning and end of a string.

print("Column names after cleaning:")
print(df.columns)
```

### Step 2: Check and Remove Duplicate Rows

We count duplicate rows first, then remove them if any are found.

```python
# Step 2: check and remove duplicate rows.
duplicates_before = df.duplicated().sum()
print("Duplicate rows before handling:", duplicates_before)

if duplicates_before > 0:
    df = df.drop_duplicates().copy()
    print("Duplicate rows removed.")
else:
    print("No duplicate rows found.")
```

### Step 3: Handle Missing Values

We first inspect missing values, then fill numeric columns with the median and text columns with `Unknown`.

```python
# Step 3: inspect missing values before filling them.
missing_before = df.isnull().sum()
print("Missing values before handling:")
print(missing_before[missing_before > 0] if missing_before.any() else "No missing values found.")

# Fill numeric columns with the median value.
for column in df.select_dtypes(include=["number"]).columns:
    df[column] = df[column].fillna(df[column].median())

# Fill text columns with a safe label.
for column in ["Region", "Frequency", "Area"]:
    df[column] = df[column].fillna("Unknown").astype(str).str.strip()

print("Missing values after handling:")
display(df.isnull().sum())
```

- Date column have still missing value 

### Step 4: Convert the Date Column

The date column is converted to a real datetime format so we can work with months and years. Any missing or invalid date is removed here.

```python
# Step 4: convert the date column and remove invalid dates.
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")
df = df.dropna(subset=["Date"]).copy()

print("Date range:", df["Date"].min(), "to", df["Date"].max())
print("Missing values after date cleaning:")
display(df.isnull().sum())
```

### Step 5: Create Helper Features

These extra columns make the later analysis easier to read and understand.

```python
# Step 5: create helper columns for time and category analysis.
df["Month"] = df["Date"].dt.to_period("M").dt.to_timestamp()
df["Year"] = df["Date"].dt.year
df["Month_Name"] = df["Date"].dt.strftime("%B")

df["Unemployment_Category"] = pd.cut(
    df["Estimated Unemployment Rate (%)"],
    bins=[-0.01, 4, 8, float("inf")],
    labels=["Low", "Moderate", "High"],
)

print("Cleaned dataset shape:", df.shape)
display(df.head())
df.info()
```

## Region-wise Average Unemployment Rate

This view ranks the regions by their average unemployment rate so the strongest and weakest performers are easy to spot.


```python
# Identify the regions with the highest average unemployment rates.
region_unemployment = (
    df.groupby("Region")["Estimated Unemployment Rate (%)"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12, 6))
sns.barplot(x=region_unemployment.values, y=region_unemployment.index, color="steelblue")
plt.title("Top 10 Regions by Average Unemployment Rate")
plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("Region")
print("Missing values after handling numeric and text columns:")
display(df.isnull().sum())

display(region_unemployment.to_frame(name="Average Unemployment Rate (%)").round(2))
```

### This Bar graph shows

This bar chart shows the top 10 regions with the highest average unemployment rate. The longer the bar, the higher the average unemployment rate in that region. It helps us quickly see which regions are facing the most unemployment pressure.


## Area-wise Comparison

This section checks whether unemployment behaves differently in rural and urban observations.


```python
# Compare the unemployment rate distribution for rural and urban areas.
area_summary = df.groupby("Area")["Estimated Unemployment Rate (%)"].mean().sort_values()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x="Area", y="Estimated Unemployment Rate (%)", color="lightsteelblue", ax=axes[0])
axes[0].set_title("Unemployment Rate Distribution by Area")
axes[0].set_xlabel("Area")
axes[0].set_ylabel("Unemployment Rate (%)")

sns.barplot(x=area_summary.index, y=area_summary.values, color="teal", ax=axes[1])
axes[1].set_title("Average Unemployment Rate by Area")
axes[1].set_xlabel("Area")
axes[1].set_ylabel("Average Unemployment Rate (%)")

plt.tight_layout()
plt.show()
```

- ### This graph shows

The boxplot compares the spread of unemployment rates in rural and urban areas. The bar chart beside it shows the average unemployment rate for each area. Together, they tell us whether unemployment is generally higher in rural or urban observations.


## Monthly Trend Analysis

Here we convert the date column into month-level observations and look at how the unemployment rate changes over time.


```python
# Convert the cleaned date column into a month-level trend for easier comparison.
monthly_trend = (
    df.groupby(df["Date"].dt.to_period("M"))["Estimated Unemployment Rate (%)"]
      .mean()
      .reset_index()
)
monthly_trend["Date"] = monthly_trend["Date"].dt.to_timestamp()

plt.figure(figsize=(14, 6))
sns.lineplot(
    data=monthly_trend,
    x="Date",
    y="Estimated Unemployment Rate (%)",
    marker="o",
    color="darkred",
)
plt.title("Monthly Trend of Average Unemployment Rate")
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- ### This line graph shows

This line graph shows how the average unemployment rate changed from month to month. It helps us spot rises and falls over time, and it shows that unemployment increased sharply toward the later months of the dataset.


## Correlation Analysis

This section checks how the main numeric variables move together.

- A positive value means the variables generally move in the same direction.
- A negative value means they move in opposite directions.
- A value close to zero suggests a weak linear relationship.


```python
# Build a simple correlation view for the numeric columns.
numeric_columns = [
    "Estimated Unemployment Rate (%)",
    "Estimated Employed",
    "Estimated Labour Participation Rate (%)",
]

correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Heatmap of Numeric Variables")
plt.tight_layout()
plt.show()
```

- ### This correlation_matrix graph shows

This heatmap shows the relationship between the numeric columns. The color and the numbers tell us whether two variables move together or move in opposite directions. It is useful for checking whether unemployment has a strong or weak relationship with employment and labour participation.
